```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.cross_validation as CV
from sklearn.feature_extraction import DictVectorizer as DV
from datetime import date
from sklearn import linear_model as LM
import seaborn as sns
%matplotlib inline
```

```python
# read data
data = pd.read_csv('train.csv',index_col='Id')
```

```python
num_features = data.select_dtypes(exclude=['object']).columns;
obj_features = data.select_dtypes(include=['object']).columns;
```

```python
for feature in obj_features:
    plt.figure()
    sns.barplot(x=feature, y="SalePrice", data=data); 
```

```python
plt.figure(figsize = [10,10])
plt.subplot(211)
sns.residplot(x="OverallQual", y="SalePrice", data=data); 
plt.subplot(212)
sns.regplot(x="OverallQual", y="SalePrice", data=data); 
```

```python
data["OverallQualLog"]=np.log(data["OverallQual"]*10)
plt.figure(figsize = [10,10])
plt.subplot(211)
sns.residplot(x="OverallQualLog", y="SalePrice", data=data);
plt.subplot(212)
sns.regplot(x="OverallQualLog", y="SalePrice", data=data); 
```

```python
data["OverallQualSqrt"]=np.power(data["OverallQual"],0.3)
sns.residplot(x="OverallQualSqrt", y="SalePrice", data=data); 
sns.lmplot(x="OverallQualSqrt", y="SalePrice", data=data); 
```
